```python
from google_play_scraper import reviews_all, Sort
import pandas as pd
from datetime import datetime
import os
from shutil import move
from IPython.display import display
```

```python
APP_ID = "co.benx.weverse"
COUNTRY = "us"
TARGET_COUNT = 5000000  # мөңмў… лӘ©н‘ң лҰ¬л·° мҲҳ
EXCEL_FILE = "playstore_reviews.xlsx"

# вң… лӮ м§ң н•„н„° ON/OFF м„Өм •
USE_DATE_FILTER = False  # True вҶ’ лӮ м§ң н•„н„° м Ғмҡ©, False вҶ’ м Ғмҡ© м•Ҳ н•Ё
DATE_FILTER = "2025-4-08"  # м Ғмҡ©н•  лӮ м§ң

if USE_DATE_FILTER:
    filter_date = datetime.strptime(DATE_FILTER, "%Y-%m-%d")
    print(f"[INFO] {DATE_FILTER} мқҙнӣ„ мһ‘м„ұлҗң лҰ¬л·°л§Ң мҲҳм§‘н•©лӢҲлӢӨ.")
else:
    print(f"[INFO] лӮ м§ң н•„н„° мӮ¬мҡ©н•ҳм§Җ м•ҠмҠөлӢҲлӢӨ. м „мІҙ лҰ¬л·° мӮ¬мҡ©н•©лӢҲлӢӨ.")
```

```python
from tqdm import tqdm
from google_play_scraper import reviews, Sort
import pandas as pd

print(f"[INFO] м „мІҙ лҰ¬л·° мҲҳм§‘ мӢңмһ‘...")

all_reviews = []
page = 1
batch_size = 100
token = None

with tqdm(total=TARGET_COUNT, desc="лҰ¬л·° мҲҳм§‘ м§„н–үмӨ‘", unit="к°ң") as pbar:
    while len(all_reviews) < TARGET_COUNT:
        result, continuation = reviews(
            APP_ID,
            lang=COUNTRY,
            country=COUNTRY,
            sort=Sort.NEWEST,
            count=batch_size,
            continuation_token=token
        )

        if not result:
            break

        for review in result:
            review_date = review['at']
            if USE_DATE_FILTER:
                if review_date >= filter_date:
                    all_reviews.append(review)
            else:
                all_reviews.append(review)

        token = continuation
        pbar.update(len(result))
        page += 1

print(f"[INFO] лҰ¬л·° мҲҳм§‘ мҷ„лЈҢ! мҙқ {len(all_reviews)}к°ң")

# ==========================
# рҹ“Ұ DataFrame ліҖнҷҳ
# ==========================
df = pd.DataFrame(all_reviews)

# ==========================
# рҹ”Ғ мӨ‘ліө м ңкұ° (userName + at кё°мӨҖ)
# ==========================
before = len(df)
df.drop_duplicates(subset=["userName", "at"], inplace=True)
after = len(df)
print(f"[INFO] мӨ‘ліө м ңкұ° мҷ„лЈҢ: {before} вҶ’ {after}к°ң")

# ==========================
# рҹ§№ н•„мҡ” м—ҶлҠ” м»¬лҹј м ңкұ°
# ==========================
columns_to_drop = ["userImage", "reviewCreatedVersion", "replyContent", "repliedAt", "appVersion"]
existing_columns = [col for col in columns_to_drop if col in df.columns]
df.drop(columns=existing_columns, inplace=True)
print(f"[INFO] н•„мҡ” м—ҶлҠ” м»¬лҹј м ңкұ° мҷ„лЈҢ: {existing_columns}")

# ==========================
# рҹҺҜ лӘ©н‘ң к°ңмҲҳ л§һм¶”кё°
# ==========================
if after >= TARGET_COUNT:
    df = df.iloc[:TARGET_COUNT]
    print("лҰ¬л·° нҒ¬лЎӨл§Ғ мҷ„лЈҢ!")
else:
    print(f"[вҡ пёҸ WARNING] лӘ©н‘ң к°ңмҲҳм—җ лҸ„лӢ¬н•ҳм§Җ лӘ»н–ҲмҠөлӢҲлӢӨ. мөңмў… {after}к°ң лҰ¬л·°л§Ң нҷ•ліҙ.")

rename_map = {
    "content": "comment",
    "score": "score",
    "at": "date",
    "userName": "username",
    "thumbsUpCount": "likes count"
}
df.rename(columns=rename_map, inplace=True)

df.to_excel(EXCEL_FILE, index=False)
print(f"[вң… м ҖмһҘ мҷ„лЈҢ] '{EXCEL_FILE}'м—җ мөңмў… {len(df)}к°ң лҰ¬л·° м ҖмһҘлҗЁ!")

from google.colab import files
files.download(EXCEL_FILE)
```

```python
import pandas as pd
import re
import emoji
from tqdm import tqdm
from google.colab import files

# нҢҢмқј м—…лЎңл“ң
uploaded = files.upload()
file_name = list(uploaded.keys())[0]

# tqdmкіј pandas м—°лҸҷ
tqdm.pandas()

# м—‘м…Җ нҢҢмқј мқҪкё°
df = pd.read_excel(file_name)

# мқҙлӘЁм§Җ + л№„мҳҒм–ҙ л¬ёмһҗ м ңкұ° н•ЁмҲҳ
def clean_text(text):
    if isinstance(text, float):
        text = str(text)
    text = emoji.replace_emoji(text, replace="")
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text.strip()

# м „мІҳлҰ¬ м Ғмҡ©
df['cleaned_comment'] = df['comment'].progress_apply(clean_text)

# xлӢЁм–ҙ мқҙн•ҳ лҰ¬л·° м ңкұ°
df['word_count'] = df['cleaned_comment'].apply(lambda x: len(x.split()))
df = df[df['word_count'] > 5].reset_index(drop=True)

# рҹ”„ м—ҙ мқҙлҰ„ ліҖкІҪ
df = df.rename(columns={
    'comment': 'previous comment',        # мӣҗлһҳ comment вҶ’ previous comment
    'cleaned_comment': 'comment'          # м „мІҳлҰ¬лҗң comment вҶ’ comment
})

# рҹ‘ү word_count м—ҙмқҖ л¶Ҳн•„мҡ”н•ҳл©ҙ м ңкұ°
df = df.drop(columns=['word_count'])

# вң… м—‘м…ҖлЎң м ҖмһҘ
preprocessed_file_path = 'preprocessed_comments.xlsx'
df.to_excel(preprocessed_file_path, index=False)
print(f"вң… м „мІҳлҰ¬лҗң лҚ°мқҙн„°лҘј '{preprocessed_file_path}' нҢҢмқјлЎң м ҖмһҘн–ҲмҠөлӢҲлӢӨ.")

# лӢӨмҡҙлЎңл“ң
files.download(preprocessed_file_path)
```

#4. BIG 5 Personality л¬ёмһҘ кҙҖл Ём„ұ мёЎм •(Zero-shot classification)

OPENNESS

```python
# рҹ“ҡ лқјмқҙлёҢлҹ¬лҰ¬ л¶Ҳлҹ¬мҳӨкё°
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
from google.colab import files

# в¬ҶпёҸ м—‘м…Җ нҢҢмқј м—…лЎңл“ң
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# рҹ“„ м—‘м…Җ лҚ°мқҙн„° л¶Ҳлҹ¬мҳӨкё°
df = pd.read_excel(filename)

# вң… 'comment' м—ҙ мЎҙмһ¬ нҷ•мқё
if 'comment' not in df.columns:
    raise ValueError("вқҢ 'comment' м—ҙмқҙ м—‘м…Җм—җ м—ҶмҠөлӢҲлӢӨ!")

# рҹӨ– Zero-shot л¶„лҘҳкё° лЎңл“ң
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

# рҹҺҜ Openness лқјлІЁ м •мқҳ
openness_labels = [
    ("Very Conservative", 0.0),
    ("A bit Conservative", 0.25),
    ("Neutral", 0.5),
    ("A bit Open", 0.75),
    ("Very Open", 1.0)
]

# рҹ§  Opennessл§Ң softmax кё°л°ҳ м җмҲҳ кі„мӮ° н•ЁмҲҳ
def classify_openness_score(comment):
    if pd.isna(comment) or not str(comment).strip():
        return 0.0

    labels = [label for label, _ in openness_labels]

    output = classifier(
        comment,
        candidate_labels=labels,
        hypothesis_template="The personality of the respondent is {} in terms of Big Five Factors.",
        return_all_scores=True
    )

    # outputмқҖ list of dicts нҳ•нғңм—¬м•ј н•Ё
    if isinstance(output, list):
        result = output
    elif isinstance(output, dict) and "scores" in output:
        result = [
            {'label': label, 'score': score}
            for label, score in zip(output["labels"], output["scores"])
        ]
    else:
        raise ValueError("лӘЁлҚё м¶ңл Ҙ нҳ•мӢқмқҙ мҳҲмғҒкіј лӢӨлҰ…лӢҲлӢӨ")

    weighted = sum(r['score'] * w for r, (_, w) in zip(result, openness_labels))
    return round(weighted, 4)

# рҹҸғ л¶„м„қ мӢӨн–ү
tqdm.pandas()
df["Openness_score"] = df["comment"].progress_apply(classify_openness_score)

# рҹ’ҫ кІ°кіј м ҖмһҘ л°Ҹ лӢӨмҡҙлЎңл“ң
output_file = filename.replace(".xlsx", "_openness_scored.xlsx")
df.to_excel(output_file, index=False)
files.download(output_file)
```

#8. к°җм„ұ л¶„м„қ(RoBERTa)

```python
# рҹ“ҡ лқјмқҙлёҢлҹ¬лҰ¬
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
from tqdm import tqdm
from google.colab import files

# вң… л””л°”мқҙмҠӨ м„Өм • (GPU мһҲмңјл©ҙ мӮ¬мҡ©)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# рҹ“Ӯ нҢҢмқј м—…лЎңл“ң
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# рҹ“„ лҚ°мқҙн„° лЎңл”©
df = pd.read_excel(filename)
if 'comment' not in df.columns:
    raise ValueError("вқҢ 'comment' м—ҙмқҙ м—ҶмҠөлӢҲлӢӨ.")
df['comment'] = df['comment'].fillna("")

# рҹӨ– лӘЁлҚё лЎңл“ң
MODEL = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL).to(device)

# вң… к°җм„ұ л¶„м„қ н•ЁмҲҳ (batch мІҳлҰ¬)
def analyze_sentiment_batch(texts):
    inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True, max_length=128)
    inputs = {k: v.to(device) for k, v in inputs.items()}
    with torch.no_grad():
        outputs = model(**inputs)
    probs = softmax(outputs.logits.cpu().numpy(), axis=1)
    return probs  # shape: (batch_size, 3)

# вң… л°°м№ҳ к°җм„ұ л¶„м„қ мӢӨн–ү
batch_size = 32
scores = []

for i in tqdm(range(0, len(df), batch_size)):
    batch_texts = df['comment'].iloc[i:i+batch_size].tolist()
    try:
        probs = analyze_sentiment_batch(batch_texts)
        for prob in probs:
            pos = prob[2]
            neg = prob[0]
            neu = prob[1]
            # compound кі„мӮ°мӢқ: (pos - neg) * (1 - neu)
            compound = round((pos - neg) * (1 - neu), 4)
            scores.append(compound)
    except Exception as e:
        print(f"вқҢ мҳӨлҘҳ л°ңмғқ at batch {i}: {e}")
        scores.extend([0.0] * len(batch_texts))

# вң… кІ°кіј м ҖмһҘ
df['sentiment_score'] = scores
output_file = filename.replace(".xlsx", "_roberta_sentiment_score_batch.xlsx")
df.to_excel(output_file, index=False)
files.download(output_file)

print("вң… л°°м№ҳ кё°л°ҳ RoBERTa к°җм„ұ л¶„м„қ мҷ„лЈҢ! м •нҷ•н•ҳкі  л№ лҘҙкІҢ мІҳлҰ¬лҗҗмҠөлӢҲлӢӨ.")
```

```python
# рҹ“ҡ лқјмқҙлёҢлҹ¬лҰ¬ л¶Ҳлҹ¬мҳӨкё°
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
from google.colab import files

# вң… м—‘м…Җ нҢҢмқј м—…лЎңл“ң
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# вң… м—‘м…Җ лҚ°мқҙн„° лЎңл”©
df = pd.read_excel(filename)

# вң… м»¬лҹј нҷ•мқё
print("\nвң… лҚ°мқҙн„°м…Ӣ м»¬лҹј лӘ©лЎқ:")
for idx, col in enumerate(df.columns):
    print(f"{idx}: {col}")

# вң… мӮ¬мҡ©мһҗ мһ…л Ҙ: мў…мҶҚ ліҖмҲҳ
target_column = input("\nрҹҺҜ мў…мҶҚліҖмҲҳлЎң мӮ¬мҡ©н•  м»¬лҹјлӘ…мқ„ мһ…л Ҙн•ҳм„ёмҡ”: ")

# вң… мӮ¬мҡ©мһҗ мһ…л Ҙ: лҸ…лҰҪ ліҖмҲҳ
input_columns = input("\nрҹ“Ң лҸ…лҰҪліҖмҲҳлЎң мӮ¬мҡ©н•  м»¬лҹјлӘ…мқ„ мүјн‘ң(,)лЎң кө¬л¶„н•ҙм„ң мһ…л Ҙн•ҳм„ёмҡ”:\nрҹ‘ү ")
independent_columns = [col.strip() for col in input_columns.split(",")]

# вң… X, y м •мқҳ
X = df[independent_columns]
y = df[target_column]

# вң… кіөл¶„мӮ° н–үл ¬ м¶ңл Ҙ
print("\nрҹ“Ҡ [кіөл¶„мӮ° н–үл ¬] (лҸ…лҰҪліҖмҲҳ к°„ мғҒкҙҖ м •лҸ„)")
print(X.cov())

# вң… мғҒкҙҖкі„мҲҳ н–үл ¬ м¶ңл Ҙ
print("\nрҹ“Ҳ [мғҒкҙҖкі„мҲҳ н–үл ¬] (лҸ…лҰҪліҖмҲҳ к°„ мғҒкҙҖ м •лҸ„)")
print(X.corr())

# вң… мғҒкҙҖкі„мҲҳк°Җ 0.7 мқҙмғҒмқё ліҖмҲҳмҢҚл§Ң м¶”м¶ң
print("\nрҹ“Ң [мғҒкҙҖкі„мҲҳ вүҘ 0.7 мқё ліҖмҲҳмҢҚ лӘ©лЎқ]")
corr_matrix = X.corr()
high_corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i + 1, len(corr_matrix.columns)):
        var1 = corr_matrix.columns[i]
        var2 = corr_matrix.columns[j]
        corr_value = corr_matrix.iloc[i, j]
        if abs(corr_value) >= 0.7:
            high_corr_pairs.append((var1, var2, corr_value))

if high_corr_pairs:
    for var1, var2, corr in high_corr_pairs:
        print(f"{var1} вҶ” {var2}: мғҒкҙҖкі„мҲҳ = {corr:.3f}")
else:
    print("вң… 0.7 мқҙмғҒмқё ліҖмҲҳмҢҚ м—ҶмқҢ")

# вң… VIF кі„мӮ°
print("\nрҹ“Ң [VIF (л¶„мӮ°нҢҪм°Ҫкі„мҲҳ) кІ°кіј]")
X_vif = sm.add_constant(X)  # м ҲнҺё нҸ¬н•Ё
vif_data = pd.DataFrame()
vif_data["Variable"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]
print(vif_data)

# вң… нҡҢк·Җ л¶„м„қ мӢӨн–ү
model = sm.OLS(y, X_vif).fit()

# вң… нҡҢк·Җ л¶„м„қ мҡ”м•Ҫ м¶ңл Ҙ
print("\nрҹ“Ҳ [нҡҢк·Җл¶„м„қ кІ°кіј мҡ”м•Ҫ]")
print(model.summary())

# вң… мң мқҳм„ұ н‘ңмӢңмҡ© лі„ мғқм„ұ н•ЁмҲҳ
def pval_to_stars(p):
    if p <= 0.001:
        return "***"
    elif p <= 0.01:
        return "**"
    elif p <= 0.05:
        return "*"
    else:
        return ""

# вң… нҡҢк·Җ кі„мҲҳ мҡ”м•Ҫн‘ң мғқм„ұ
summary_df = pd.DataFrame({
    "Variable": model.params.index,
    "Coefficient": model.params.values,
    "P-value": model.pvalues.values,
    "Significance": model.pvalues.apply(pval_to_stars)
})

# вң… м¶ңл Ҙ
print("\nрҹ“Ӣ нҡҢк·Җ кі„мҲҳ + мң мқҳм„ұ(* н‘ңмӢң) мҡ”м•Ҫн‘ң:")
print(summary_df.to_string(index=False))
print(f"\nрҹ“Ң Adjusted RВІ (мЎ°м •лҗң м„ӨлӘ…л Ҙ): {model.rsquared_adj:.4f}")
```

X -> M -> Y

```python
# рҹ“ҡ лқјмқҙлёҢлҹ¬лҰ¬ л¶Ҳлҹ¬мҳӨкё°
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from google.colab import files
from statsmodels.stats.outliers_influence import variance_inflation_factor

# вң… м—‘м…Җ нҢҢмқј м—…лЎңл“ң
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# вң… м—‘м…Җ лҚ°мқҙн„° лЎңл”©
df = pd.read_excel(filename)

# вң… м»¬лҹј лӘ©лЎқ м¶ңл Ҙ
print("\nвң… лҚ°мқҙн„°м…Ӣ м»¬лҹј лӘ©лЎқ:")
for idx, col in enumerate(df.columns):
    print(f"{idx}: {col}")

# вң… мӮ¬мҡ©мһҗ мһ…л Ҙ
target_column = input("\nрҹҺҜ мў…мҶҚліҖмҲҳлЎң мӮ¬мҡ©н•  м»¬лҹјлӘ…мқ„ мһ…л Ҙн•ҳм„ёмҡ”: ")
input_columns = input("\nрҹ“Ң лҸ…лҰҪліҖмҲҳлЎң мӮ¬мҡ©н•  м»¬лҹјлӘ…мқ„ мүјн‘ң(,)лЎң кө¬л¶„н•ҙм„ң мһ…л Ҙн•ҳм„ёмҡ”:\nрҹ‘ү ")
independent_columns = [col.strip() for col in input_columns.split(",")]
mediator_column = input("\nрҹ§© л§Өк°ңліҖмҲҳ(мӨ‘мһ¬ліҖмҲҳ)лЎң мӮ¬мҡ©н•  м»¬лҹјлӘ…мқ„ мһ…л Ҙн•ҳм„ёмҡ”:\nрҹ‘ү ")

# вң… ліҖмҲҳ м„Өм •
X = df[independent_columns]
M = df[[mediator_column]]
y = df[target_column]

# вң… кіөл¶„мӮ° н–үл ¬ м¶ңл Ҙ
print("\nрҹ“Ҡ [кіөл¶„мӮ° н–үл ¬]")
print(pd.concat([X, M, y], axis=1).cov())

# вң… н‘ңмӨҖнҷ” н•ЁмҲҳ
def standardize(df):
    return pd.DataFrame(StandardScaler().fit_transform(df), columns=df.columns)

# вң… мң мқҳм„ұ н‘ңмӢң н•ЁмҲҳ
def significance_stars(p):
    if p < 0.001:
        return '***'
    elif p < 0.01:
        return '**'
    elif p < 0.05:
        return '*'
    elif p < 0.1:
        return '.'
    else:
        return ''

def print_coef_with_significance(model):
    summary_df = pd.DataFrame({
        'Coef': model.params,
        'p-value': model.pvalues,
        'Signif': model.pvalues.apply(significance_stars)
    })
    print(summary_df)

# вң… н‘ңмӨҖнҷ”лҗң лҚ°мқҙн„°
X_std = standardize(X)
M_std = standardize(M)
y_std = standardize(y.to_frame())

# вң… 1лӢЁкі„: X вһқ y
X1_const = sm.add_constant(X)
model1 = sm.OLS(y, X1_const).fit()

X1_std_const = sm.add_constant(X_std)
model1_std = sm.OLS(y_std, X1_std_const).fit()

print("\nрҹ”№ [1лӢЁкі„: лҸ…лҰҪліҖмҲҳ вһқ мў…мҶҚліҖмҲҳ]")
print(model1.summary())
print(f"\nрҹ“Ң Adjusted RВІ: {model1.rsquared_adj:.4f}")
print("\nрҹ“Ҡ Standardized Betas with significance:")
print_coef_with_significance(model1_std)

# вң… 2лӢЁкі„: X вһқ M
model2 = sm.OLS(M, X1_const).fit()
model2_std = sm.OLS(M_std, X1_std_const).fit()

print("\nрҹ”№ [2лӢЁкі„: лҸ…лҰҪліҖмҲҳ вһқ л§Өк°ңліҖмҲҳ]")
print(model2.summary())
print(f"\nрҹ“Ң Adjusted RВІ: {model2.rsquared_adj:.4f}")
print("\nрҹ“Ҡ Standardized Betas with significance:")
print_coef_with_significance(model2_std)

# вң… 3лӢЁкі„: X + M вһқ y
XM = pd.concat([X, M], axis=1)
XM_const = sm.add_constant(XM)
model3 = sm.OLS(y, XM_const).fit()

XM_std = pd.concat([X_std, M_std], axis=1)
XM_std_const = sm.add_constant(XM_std)
model3_std = sm.OLS(y_std, XM_std_const).fit()

print("\nрҹ”№ [3лӢЁкі„: лҸ…лҰҪліҖмҲҳ + л§Өк°ңліҖмҲҳ вһқ мў…мҶҚліҖмҲҳ]")
print(model3.summary())
print(f"\nрҹ“Ң Adjusted RВІ: {model3.rsquared_adj:.4f}")
print("\nрҹ“Ҡ Standardized Betas with significance:")
print_coef_with_significance(model3_std)

# вң… VIF кі„мӮ° н•ЁмҲҳ
def compute_vif(df_with_const):
    vif_df = pd.DataFrame()
    vif_df["Variable"] = df_with_const.columns
    vif_df["VIF"] = [variance_inflation_factor(df_with_const.values, i)
                     for i in range(df_with_const.shape[1])]
    return vif_df

# вң… VIF м¶ңл Ҙ
print("\nрҹ“Ң [VIF (лӢӨмӨ‘кіөм„ м„ұ м§Җн‘ң)]")
vif_results = compute_vif(XM_const)
print(vif_results)

# вң… мң„кі„нҡҢк·Җ кІҖм • (RВІ мҰқк°Җлҹүмқҳ мң мқҳм„ұ кІҖм •)
anova_results = sm.stats.anova_lm(model1, model3)

anova_results_rounded = anova_results.copy()
anova_results_rounded['F'] = anova_results_rounded['F'].apply(lambda x: f"{x:.4f}" if pd.notnull(x) else "")
anova_results_rounded['Pr(>F)'] = anova_results_rounded['Pr(>F)'].apply(lambda x: f"{x:.4f}" if pd.notnull(x) else "")

print("\nрҹ“Ҳ [мң„кі„нҡҢк·Җ: RВІ мҰқк°Җ мң мқҳм„ұ кІҖм • (ANOVA л№„көҗ)]")
print(anova_results_rounded)
```

```python
# рҹ“ҡ лқјмқҙлёҢлҹ¬лҰ¬ л¶Ҳлҹ¬мҳӨкё°
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
from google.colab import files
from statsmodels.stats.outliers_influence import variance_inflation_factor

# вң… м—‘м…Җ нҢҢмқј м—…лЎңл“ң
uploaded = files.upload()
filename = list(uploaded.keys())[0]

# вң… м—‘м…Җ лҚ°мқҙн„° лЎңл”©
df = pd.read_excel(filename)

# вң… м»¬лҹј лӘ©лЎқ м¶ңл Ҙ
print("\nвң… лҚ°мқҙн„°м…Ӣ м»¬лҹј лӘ©лЎқ:")
for idx, col in enumerate(df.columns):
    print(f"{idx}: {col}")

# вң… мӮ¬мҡ©мһҗ мһ…л Ҙ
target_column = input("\nрҹҺҜ мў…мҶҚліҖмҲҳлЎң мӮ¬мҡ©н•  м»¬лҹјлӘ…мқ„ мһ…л Ҙн•ҳм„ёмҡ”: ")
input_columns = input("\nрҹ“Ң лҸ…лҰҪліҖмҲҳлЎң мӮ¬мҡ©н•  м»¬лҹјлӘ…мқ„ мүјн‘ң(,)лЎң кө¬л¶„н•ҙм„ң мһ…л Ҙн•ҳм„ёмҡ”:\nрҹ‘ү ")
independent_columns = [col.strip() for col in input_columns.split(",")]
mediator_column = input("\nрҹ§© л§Өк°ңліҖмҲҳ(мӨ‘мһ¬ліҖмҲҳ)лЎң мӮ¬мҡ©н•  м»¬лҹјлӘ…мқ„ мһ…л Ҙн•ҳм„ёмҡ”:\nрҹ‘ү ")

# вң… ліҖмҲҳ м„Өм •
X = df[independent_columns]
M = df[[mediator_column]]
y = df[target_column]

# вң… кіөл¶„мӮ° н–үл ¬ м¶ңл Ҙ
print("\nрҹ“Ҡ [кіөл¶„мӮ° н–үл ¬]")
print(pd.concat([X, M, y], axis=1).cov())

# вң… н‘ңмӨҖнҷ” н•ЁмҲҳ
def standardize(df):
    return pd.DataFrame(StandardScaler().fit_transform(df), columns=df.columns)

# вң… мң мқҳм„ұ н‘ңмӢң н•ЁмҲҳ
def significance_stars(p):
    if p < 0.001:
        return '***'
    elif p < 0.01:
        return '**'
    elif p < 0.05:
        return '*'
    elif p < 0.1:
        return '.'
    else:
        return ''

def print_coef_with_significance(model):
    summary_df = pd.DataFrame({
        'Coef': model.params,
        'p-value': model.pvalues,
        'Signif': model.pvalues.apply(significance_stars)
    })
    print(summary_df)

# вң… н‘ңмӨҖнҷ”лҗң лҚ°мқҙн„°
X_std = standardize(X)
M_std = standardize(M)
y_std = standardize(y.to_frame())

# вң… 1лӢЁкі„: X вһқ y
X1_const = sm.add_constant(X)
model1 = sm.OLS(y, X1_const).fit()

X1_std_const = sm.add_constant(X_std)
model1_std = sm.OLS(y_std, X1_std_const).fit()

print("\nрҹ”№ [1лӢЁкі„: лҸ…лҰҪліҖмҲҳ вһқ мў…мҶҚліҖмҲҳ]")
print(model1.summary())
print(f"\nрҹ“Ң Adjusted RВІ: {model1.rsquared_adj:.4f}")
print("\nрҹ“Ҡ Standardized Betas with significance:")
print_coef_with_significance(model1_std)

# вң… 2лӢЁкі„: X вһқ M
model2 = sm.OLS(M, X1_const).fit()
model2_std = sm.OLS(M_std, X1_std_const).fit()

print("\nрҹ”№ [2лӢЁкі„: лҸ…лҰҪліҖмҲҳ вһқ л§Өк°ңліҖмҲҳ]")
print(model2.summary())
print(f"\nрҹ“Ң Adjusted RВІ: {model2.rsquared_adj:.4f}")
print("\nрҹ“Ҡ Standardized Betas with significance:")
print_coef_with_significance(model2_std)

# вң… 2лӢЁкі„ VIF м¶ңл Ҙ
def compute_vif(df_with_const):
    vif_df = pd.DataFrame()
    vif_df["Variable"] = df_with_const.columns
    vif_df["VIF"] = [variance_inflation_factor(df_with_const.values, i)
                     for i in range(df_with_const.shape[1])]
    return vif_df

print("\nрҹ“Ң [2лӢЁкі„ VIF (лӢӨмӨ‘кіөм„ м„ұ м§Җн‘ң)]")
vif_results_2nd = compute_vif(X1_const)
print(vif_results_2nd)

# вң… 3лӢЁкі„: X + M вһқ y
XM = pd.concat([X, M], axis=1)
XM_const = sm.add_constant(XM)
model3 = sm.OLS(y, XM_const).fit()

XM_std = pd.concat([X_std, M_std], axis=1)
XM_std_const = sm.add_constant(XM_std)
model3_std = sm.OLS(y_std, XM_std_const).fit()

print("\nрҹ”№ [3лӢЁкі„: лҸ…лҰҪліҖмҲҳ + л§Өк°ңліҖмҲҳ вһқ мў…мҶҚліҖмҲҳ]")
print(model3.summary())
print(f"\nрҹ“Ң Adjusted RВІ: {model3.rsquared_adj:.4f}")
print("\nрҹ“Ҡ Standardized Betas with significance:")
print_coef_with_significance(model3_std)

# вң… 3лӢЁкі„ VIF м¶ңл Ҙ
print("\nрҹ“Ң [VIF (лӢӨмӨ‘кіөм„ м„ұ м§Җн‘ң)]")
vif_results = compute_vif(XM_const)
print(vif_results)

# вң… мң„кі„нҡҢк·Җ кІҖм • (RВІ мҰқк°Җлҹүмқҳ мң мқҳм„ұ кІҖм •)
anova_results = sm.stats.anova_lm(model1, model3)

anova_results_rounded = anova_results.copy()
anova_results_rounded['F'] = anova_results_rounded['F'].apply(lambda x: f"{x:.4f}" if pd.notnull(x) else "")
anova_results_rounded['Pr(>F)'] = anova_results_rounded['Pr(>F)'].apply(lambda x: f"{x:.4f}" if pd.notnull(x) else "")

print("\nрҹ“Ҳ [мң„кі„нҡҢк·Җ: RВІ мҰқк°Җ мң мқҳм„ұ кІҖм • (ANOVA л№„көҗ)]")
print(anova_results_rounded)
```
